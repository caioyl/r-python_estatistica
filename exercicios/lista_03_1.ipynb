````python
# Para descobrir o coeficiente angular e linear de uma reta ajustada a partir de um DataFrame com duas variáveis no Pandas, você pode usar a biblioteca NumPy para realizar a regressão linear. Aqui está um exemplo de como fazer isso:

# ```python
import pandas as pd
import numpy as np

# Realizar a regressão linear usando numpy
x = df['Dosagem (x)']
y = df['Proporção de pacientes (%) Y']
coefficients = np.polyfit(x, y, 1)  # O último argumento é a ordem do polinômio (1 para regressão linear)

coef_angular = coefficients[0]
coef_linear = coefficients[1]

print("Coeficiente Angular:", coef_angular)
print("Coeficiente Linear:", coef_linear)
# ```



#Lembre-se de que a regressão linear assume que a relação entre as variáveis é aproximadamente linear. Portanto, esse método pode não ser adequado se a relação for mais complexa. Além disso, considere outras bibliotecas como o `scipy.stats.linregress` para obter resultados semelhantes.
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Separando as variáveis independentes (X) e dependentes (Y)
X = df[['Dosagem (x)']]
y = df['Proporção de pacientes (%) Y']

# Inicializando o modelo de regressão linear
model = LinearRegression()

# Treinando o modelo
model.fit(X, y)

# Coeficientes da reta de regressão
slope = model.coef_[0]
intercept = model.intercept_

# Criando um array de valores X para a reta de regressão
x_range = np.linspace(df['Dosagem (x)'].min(), df['Dosagem (x)'].max(), num=100)

# Calculando os valores correspondentes de Y usando a equação da reta de regressão
y_range = slope * x_range + intercept

# Criando o gráfico de dispersão linear com a reta de regressão
plt.scatter(df['Dosagem (x)'], df['Proporção de pacientes (%) Y'], label='Dados')
plt.plot(x_range, y_range, color='red', label='Reta de Regressão')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Gráfico de Dispersão Linear com Reta de Regressão')
plt.show()
````

````python
lista_y = []
def valor_y(x):
  y = 10.3 + 0.116 * x
  return y

print(len(df['Proporção de pacientes (%) Y']))

for c in range(len(df['Proporção de pacientes (%) Y'])):
  x = df['Dosagem (x)'][c]
  lista_y.append(valor_y(x))

df['Valores_estimados'] = lista_y
df['Diferença'] = df['Valores_estimados'] - df['Proporção de pacientes (%) Y']

df.to_csv('diferenças.csv',index=False)
````
